```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os.path import dirname, join

from bokeh.plotting import figure
from bokeh.io import show, output_notebook, output_file, push_notebook, curdoc
from bokeh.client import push_session, pull_session

from bokeh.models import ColumnDataSource, HoverTool, Legend, Panel, CustomJS
from bokeh.models.widgets import CheckboxGroup, Slider, RadioGroup, Tabs

from bokeh.layouts import column, row, WidgetBox
from bokeh.palettes import Category20_16
```

```python
src_csv = pd.read_csv(join(dirname('__file__'), './', 'data', 'NPS.tsv'), na_values=['-9', '-2', '-1', '-8'], sep='\t')
```

```python
def make_dataset(variables, year = 2000):
    plot_src = src_csv.query('STATEID < 60 and STATEID != 52 and STATEID != 11')
    variables.append('STATE')
    plot_src = new_src.loc[new_src['YEAR'] == year, variables]
    plot_src = new_src.set_index('STATE')
    return ColumnDataSource(plot_src) 
    
# draw the plot
def make_plot(plot_src, year):
    bp = figure(plot_width = 1200, plot_height = 600,
                title = "Demographics Per State, " + str(year),
                x_range = plot_src.data['STATE'],
                x_axis_label = 'State', y_axis_label = 'Number of Prisoners')

    states = plot_src.data['STATE']
    races = list()
    colors = list()
    for i, x in enumerate(plot_src.data):
        if x == 'index' or x == 'STATE':
            continue
        races.append(x)
        colors.append(Category20_16[i])

    races_legend = list()
    for name in races:
        if name == 'ASIANM':
            races_legend.append('Asian, Male') 
        elif name == 'ASIANF':
            races_legend.append('Asian, Female')
        elif name == 'WHITEF':
            races_legend.append('White, Female')
        elif name == 'WHITEM':
            races_legend.append('White, Male')
        elif name == 'BLACKF':
            races_legend.append('Black, Female')
        elif name == 'BLACKM':
            races_legend.append('Black, Male')
        elif name == 'HISPM':
            races_legend.append('Hispanic/Latino, Male')
        elif name == 'HISF':
            races_legend.append('Hispanic/Latino, Female')

    AM = bp.vbar_stack(races, x = 'STATE', width = 0.9, color = colors, source = plot_src.data,
                           legend = races_legend)
    
    h = HoverTool(tooltips = [
        ('State', '@STATE'),
        ('# of Prisoners', '@races')
        ], names = races)

    bp.add_tools(h)

    bp.legend.location = 'top_right'
    bp.legend.click_policy='hide'
    
    return bp
    
    # update function
def update(attr, old, new):
    # get list of variables to update
    demo_to_plot = [demo_selection.labels[i] for i in 
                            demo_selection.active]
        
    # update the slider for year
    year = year_select.value

    # update the dataset
    new_src = make_dataset(demo_to_plot, year)
    src.data.update(new_src.data)
    print(src.data)
     
#create a checkbox
race = ['ASIANM', 'ASIANF', 'WHITEM', 'WHITEF', 'BLACKM', 'BLACKF', 'HISPM', 'HISPF']
demo_selection =  CheckboxGroup(labels = race, active=[0, 1])
demo_selection.on_change('active', update)

    # create a slider for year
year_select = Slider(start = 1978, end = 2016, step = 1, value = 2001, title = 'Year')
year_select.on_change('value', update)
    
    # Initialize the plot
initial_active = [demo_selection.labels[i] for i in demo_selection.active]
print(initial_active)
print(year_select.value)


src = make_dataset(initial_active, year_select.value)
p = make_plot(src, year_select.value)


    # add the controls
controls = WidgetBox(year_select, demo_selection)
    
    # create a layout
layout = row(controls, p)
    
    # add a tab
tab1 = Panel(child = layout, title = 'Prison Admissions by Ethnicity')
```

```python
tabs = Tabs(tabs = [tab1])
curdoc().add_root(tabs)

show(layout)
output_notebook()
```
